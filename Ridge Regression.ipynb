```python
def ridge_regression(X,Y, learning_rate, epochs, l2_penalty):
    w = 0
    b = 0

    samples = float(len(X))

    # Performing Gradient Descent 
    for i in range(epochs): 
        Y_pred = w*X + b
        
        cost = (1/(2 * samples)) * sum((Y - Y_pred) * ((Y - Y_pred))) + (l2_penalty/2) * (w*w)
        dw = ((-2 * sum((X * (Y - Y_pred))) + (2 * l2_penalty * w))) / samples
        db = (-2 * sum(Y - Y_pred)) / samples
        
        #Updating Coefficients
        w = w - learning_rate * dw  
        b = b - learning_rate * db 

    return (w, b, cost)
```

```python
def iris():
    import numpy as np
    import pandas as pd
    from sklearn.model_selection import train_test_split
    import matplotlib.pyplot as plt
    df = pd.read_csv("iris.csv")
    
    X = df.iloc[:,1].values #sepal length
    Y = df.iloc[:,2].values #sepal width
    
    l2 = 0.00001
    lr = 0.0001
    
    w,b,cost = ridge_regression(X,Y, learning_rate = lr, epochs = 1000, l2_penalty = l2)
    Y_pred = w*X + b
    
    #Plotting regression graph
    plt.title("Sepal Length vs Sepal Width")
    plt.scatter(X, Y) 
    plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
    plt.xlabel("Sepal Length")
    plt.ylabel("Sepal Width")
    plt.show()
    
        
    plot_l2_v_cost(X,Y,l2)
    plot_learning_rate_v_cost(X,Y,lr)
    plot_epochs_v_cost(X,Y)
```

```python
iris()
```

```python
def plotss(X, Y, plot_label, label_x, label_y):
    import matplotlib.pyplot as plt
    
    plt.title(plot_label)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.scatter(X, Y)
    plt.plot(X, Y, color = 'red')
    plt.show()
```

```python
def car_price():
    import numpy as np
    import pandas as pd
    from sklearn.model_selection import train_test_split
    import matplotlib.pyplot as plt
    df = pd.read_csv("CarPrice_Assignment.csv")
    
    X = df.iloc[:,19].values 
    Y = df.iloc[:,24].values
    
#     print(Y)
    l2 = 0.00001
    lr = 0.0001
    
    w,b,cost = ridge_regression(X,Y, learning_rate = lr, epochs = 1000, l2_penalty = l2)
    Y_pred = w*X + b
    
#   Plotting regression graph
    plt.title("Stroke vs Highway Mileage")
    plt.scatter(X, Y) 
    plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
    plt.xlabel("Strokes")
    plt.ylabel("Highway Mileage")
    plt.show()
    
        
    plot_l2_v_cost(X,Y,l2)
    plot_learning_rate_v_cost(X,Y,lr)
    plot_epochs_v_cost(X,Y)
```

```python
def plot_l2_v_cost(X,Y,l2):
    costs =[]
    l2_values = []
    for _ in range (30):
        
        w,b,cost = ridge_regression(X,Y, learning_rate = 0.001, epochs = 200, l2_penalty = l2)
        Y_pred = w*X + b

        costs.append(cost)
        l2_values.append(l2)
        l2 = l2*2
        
    plotss(X = l2_values, Y = costs, plot_label = "L2 Value vs Cost", label_x = "L2", label_y = "Cost")
```

```python
def plot_learning_rate_v_cost(X,Y,lr):
    costs =[]
    lr_values = []
    for _ in range (10):
        
        w,b,cost = ridge_regression(X,Y, learning_rate = lr, epochs = 200, l2_penalty = 0.0001)
        Y_pred = w*X + b

        costs.append(cost)
        lr_values.append(lr)
        lr = lr*2
        
    plotss(X = lr_values, Y = costs, plot_label = "Learning Rate vs Cost", label_x = "Learning Rate", label_y = "Cost")
```

```python
def plot_epochs_v_cost(X,Y):
    costs =[]
    epoch_values = []
    epoch = 2
    for _ in range (10):
        
        w,b,cost = ridge_regression(X,Y, learning_rate = 0.001, epochs = epoch, l2_penalty = 0.0001)
        Y_pred = w*X + b

        costs.append(cost)
        epoch_values.append(epoch)
        epoch = epoch*2
        
    plotss(X = epoch_values, Y = costs, plot_label = "Epochs vs Cost", label_x = "Epochs", label_y = "Cost")
```
